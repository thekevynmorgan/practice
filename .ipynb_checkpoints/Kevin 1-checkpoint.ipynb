```python
def square (x):
    return x * x

def sum_of_square(x, y):
    return square(x) + square(y)

print(sum_of_square(2,3))
```

```python
def square(x):
    return x * x

def cube(x):
    return x * x * x

f = square
f = cube
def fxy(f, x, y):
    return f(x) + f(y)

print(f(4))
print(fxy(square, 2, 3))
```

```python
def increment(x, amount=1):
    return x + amount

print(increment(10))
print(increment(10,5))
print(increment(10, amount=2))
```

```python
def difference(x, y):
    return x - y

print(difference(5,2))
print(difference(x=5, y=2))
print(difference(5, y=2))
print(difference(y=2, x=5))
```
