### Trees - An Overview
In a tree, the ideas of predecessor and successor (next and previous in doubly-linked list) are replaced with those of parent and child

Trees have two main characteristics:
1. Each item can have multiple children
2. All items, except a special item called the **root** have exactly one parent.

### Terminology

| Term | Definition |
| ---  | ---        |
| Node | An item stored in the tree|
| Root node | The top most node - with no parent |
| Child | A node connected to a parent. Organised in left-to-right order |
| Parent | A node which has child nodes directly connected |
| Siblings | All of the children of a common parent |
| Leaf | A node that has no children |
| Height | (or Depth) The number of edges from Root to deepest Leaf |
| Descendants | Children, childrens' children, etc. To the deepest Leaf |
| Ancestors | Parent, parents' parent, etc. to the root |
| Path | The sequence of nodes that connect a parent to one of its descendants |
| Subtree | The tree representing a node and selected descendants |

## The Binary Search Tree
Sorted collections can be represented as tree-like structures

For example, a **binary search tree** or **BST** for short can support efficient searches and insertions.

### Search Efficiency
The number of nodes, $N$, contained in a full binary tree of height $H$ is $2^{H+1} - 1$

The height, $H$, of a full binary tree with $N$ nodes is $log_2(N+1)-1$

Therefore, the operations that it takes to access a given node in a full binary tree is $O(log N)$

* A BST maintains a **sorted ordering** of nodes
    * Nodes in the left subtree of a node are < node
    * Nodes in the right subtree of a node are > node
* When the shape approaches that of a perfectly balanced binary tree, searches and insertions are $O(log(n))$ in the worst case.
* Not all BSTs are perfectly balanced, in the worst case, they become linear and support linear searches $O(n)$

### BST in Python

````python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    # Recursive insert!
    def insert(self,data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
    def find(self, data, parent = None):
        if data < self.data:
            if self.left is None:
                return None, None
            return self.left.find(data, self)
        elif data > self.data:
            if self.right is None:
                return None, None
            return self.right.find(data, self)
        else: return self, parent
        
````

### Deletion
Deletion is slightly more problematic as we may need to join an isolated subtree. We therefore need to find the node **and** its parent, hence `find()` returning both.

Cases:
* Node to be deleted is a leaf: we simply detach by setting the corresponding child pointer of the parent to `None`
* Node to be deleted has one child: the node is cut from the tree and the single child (and any subtree) is linked directly to the parent of the removed node.
* Node to be deleted has two children: There is no easy solution, so we find a minimum value in the right subtree, replace the node to be deleted with this found value, then delete the found node in the right subtree. In other words, something gets pushed higher up.

## Heaps
Heaps are a means of storing data in a tree structure in *partially sorted order* obeying a **heap condition**:
* If A is a parent node of B then the key of node A is ordered with respect to the key of node B with the same ordering applying across the heap.
* In a **max heap**, the keys of parent nodes are *always greater that or equal to* those of the children and the highest key is in the root node.
* In a **min heap**, the keys of parent nodes are *less than or equal to* those of the children and the lowest key is in the root node.

The partially sorted nature of a heap allow us to make very efficient sorting algorithms. Imagine making piles of paper on your desk, sorted by size. You place posters at one end, post-it notes at the other end, making piles of rough sizes in between.

````python
class heap():
    def __init__(self):
        self.heap=[]
    def add(self, item):
        self.heap.append(item)
        curPos = len(self.heap)-1
        while curPos>0:
            parent = (curPos - 1) // 2 # NB // Performs Integer Division
            parentItem = self.heap[parent]
            if parentItem <= item:
                break
            else:
                self.heap[curPos] = self.heap[parent]
                self.heap[parent] = item
                curPos=parent
    def pop(self):
        # Remove the top item replacing with bottom item.
        # Move down the tree swapping the minimum child value
        #   with the current node value
        #   until the heap condition is restored
        if len(self.heap)>0:
            topItem = self.heap[0]
            bottomItem = self.heap.pop(len(self.heap) - 1)
            if len(self.heap) == 0: return bottomItem
            self.heap[0] = bottomItem
            lastIndex = len(self.heap) - 1
            curPos = 0
            while True:
                leftChild = 2 * curPos + 1 
                rightChild = 2 * curPos + 2
                if leftChild > lastIndex: break
                if rightChild > lastIndex: minChild = leftChild
                else:
                    leftItem  = self.heap[leftChild]
                    rightItem = self.heap[rightChild]
                    if leftItem < rightItem: minChild = leftChild
                    else: minChild = rightChild
                if bottomItem <= self.heap[minChild]: break
                else:
                    self.heap[curPos] = self.heap[minChild]
                    self.heap[minChild] = bottomItem
                    curPos = minChild
            return topItem

h = heap()
h.add(100)
print(h.heap)
h.add(80)
print(h.heap)
h.add(70)
print(h.heap)
h.add(10)
print(h.heap)
h.add(85)
print(h.heap)
h.add(20)
print(h.heap)
h.add(50)
print(h.heap)
h.add(40)
print(h.heap)
h.add(90)
print(h.heap)
h.add(15)
print(h.heap)
````

## Graphs
### Terminology
* A **vertex** (or node) generally has a **name** (or **key**) and additional information known as the **payload**.
* An **edge** (or arc) connects two vertices to show that there is a relationship between them. Edges may be one-way or two-way. If the edges in a graph are all one-way, we say that the graph is a directed graph, or a digraph.
* Edges may be **weighted** to show that there is a cost to go from one vertex to another. For example in a graph of roads that connect one city to another, the weight on the edge might represent the distance between two cities, or it might represent the time it would take to travel between the cities.
* A **path** in a graph is a sequence of vertices that are connected by edges. The unweighted path length is the number of edges in the path. The weighted path length is the sum of the weights of all the edges in the path.
* A **cycle** in a directed graph is a path that starts and ends at the same vertex. A graph with no cycles is called an **acyclic graph**. A directed graph with no cycles is called a **directed acyclic graph** or a **DAG**.

## Abstract Data Type (ADT) for a Graph
| Method or Operator | Description |
| --- | --- |
| `Graph()` | created a new, empty graph. |
| `addVertex(vert)` | adds an instance of a vertex to the graph. |
| `addEdge(fromVert, toVert)` | adds a new, directed edge to the graph that connects two vertices. |
| `addEdge(fromVert, toVert, weight)` | adds a new, weighted, directed edge to the graph that connects two vertices. |
| `getVertex(vertKey)` | finds the vertex in the graph named `vertKey` |
| `getVertices()` | returns the list of all vertices in the graph. |
| `in` | returns `True` for a statement of the form `vertex in graph`, if the given vertex is in the graph, `False` otherwise. |

### Implementation
There are two basic approaches, the **Adjacency Matrix** and the **Adjacency List**. the former represents vertices in a 2D matrix. This approach can be wasteful, especially for graphs that are sparsely connected. A more space-efficient way to implement a sparsely connected graph is to use an adjacency list. A list of all the vertices in the `Graph` object is maintained. Each vertex object maintains a list of the vertices to which it is connected.

````python
class Vertex:
    def __init__(self, ident):
        self.ident = ident
        self.adjacent = {} # Adjacency list
    def __str__(self):
        return str(self.ident) + ' adjacent: ' + str([x.ident for x in self.adjacent])
    def addNeighbour(self, neighbour, weight=0):
        self.adjacent[neighbour] = weight
    def getConnections(self):
        return self.adjacent.keys()
    def getId(self):
        return self.ident
    def getWeight(self, neighbour):
        return self.adjacent[neighbour]

class Graph:
    def __init__(self):
        self.vertDict = {}
        self.numVertices = 0
    def __iter__(self):
        return iter(self.vertDict.values())
    def addVertex(self, id):
        self.numVertices += 1
        newVertex = Vertex(id)
        self.vertDict[id] = newVertex
        return newVertex
    def getVertex(self, n):
        if n in self.vertDict:
            return self.vertDict[n]
        else:
            return None
    def getVertices(self):
        return self.vertDict.keys()
    def addEdge(self, frm, to, cost=0):
        if frm not in self.vertDict:
            self.addVertex(frm)
        if to not in self.vertDict:
            self.addVertex(to)
        self.vertDict[frm].addNeighbour(self.vertDict[to], cost)
        self.vertDict[to].addNeighbour(self.vertDict[frm], cost)

g = Graph()
# g.addVertex('a') # not necessary
g.addEdge('a', 'b', 7)
g.addEdge('a', 'c', 9)
g.addEdge('a', 'f', 14)
g.addEdge('b', 'c', 10)
g.addEdge('b', 'd', 15)
g.addEdge('c', 'd', 11)
g.addEdge('c', 'f', 2)
g.addEdge('d', 'e', 6)
g.addEdge('e', 'f', 9)
for v in g:
    for w in v.getConnections():
        vid = v.getId()
        wid = w.getId()
        print ('( %s , %s, %3d)'  % ( vid, wid, v.getWeight(w)))
````

## Depth First Search
Depth first explores all possible vertices (from a supplied root) down each branch before backtracking. Can be implemented iteratively or recursively. For each vertex:
1. Mark the current vertex as being visited.
2. Explore each adjacent vertex that is not included in the visited set.

This implementation generally uses a stack.

## Breadth First Search
Given a graph G and a starting vertex s, a breadth first search proceeds by exploring edges in the graph to find all the vertices in G for which there is a path from s.

The remarkable thing about a breadth first search is that it finds all the vertices that are a distance k from s before it finds any vertices that are a distance k+1

One way to visualize the BFS algorithm is to imagine that it is building a tree, one level of the tree at a time. A breadth first search adds all children of the starting vertex before it begins to discover any of the grandchildren.

### tkinter Listbox Example


````python
import tkinter as tk

def clearInput(i):
    # clear passed entry
    inpt.delete(0, 50)
    
def addItem():
    # add new item to end
    lb.insert(tk.END, inpt.get())

def deleteItem():
    try:
        # get selected index
        index = lb.curselection()[0]
        lb.delete(index)
    except IndexError:
        pass
 
def getItem(event):
    # get selected text and display in entry
    index = lb.curselection()[0]
    seltext = lb.get(index)
    clearInput(inpt)
    inpt.insert(0, seltext)
def saveList():
    # get the list
    temp_list = list(lb.get(0, tk.END))
    # add a newline char to each line
    temp_list = [i + '\n' for i in temp_list]
    fout = open("file1.txt", "w")
    fout.writelines(temp_list)
    fout.close()    
def setList(event):
    try:
        index = lb.curselection()[0]
        # delete old text
        lb.delete(index)
    except IndexError:
        index = tk.END
    # insert edited item back into listbox at index
    lb.insert(index, inpt.get())
    clearInput(inpt)
def sortList():
    tList = list(lb.get(0, tk.END))
    tList.sort(key=str.lower)
    # delete contents of present listbox
    lb.delete(0, tk.END)
    # load listbox with sorted data
    for item in tList:
        lb.insert(tk.END, item)

try:
   # read the data file into a list
   fin = open("file1.txt", "r")
   slist = fin.readlines()
   fin.close() 
except:
# create initial data file
  str1 = """Brian A
Jonathan A
James C
Zhe C
Louis D
Jake D
Joseph H
Lewis M
Luke N
Joseph NdG
Shaun T"""
  fout = open("file1.txt", "w")
  fout.write(str1)
  fout.close()
  slist = list(str1.split("\n"))

# strip the trailing newline char
slist = [istr.rstrip() for istr in slist]
 
root = tk.Tk()
root.title("Listboxes")
# create the listbox with a vertical scrollbar
lb = tk.Listbox(root, width=40, height=5)
lb.grid(row=0, column=0)
yscroll = tk.Scrollbar(command=lb.yview, orient=tk.VERTICAL)
yscroll.grid(row=0, column=1, sticky=tk.N+tk.S)
lb.configure(yscrollcommand=yscroll.set)
 
# use entry to display or edit selection
inpt = tk.Entry(root, width=50, bg='cyan')
inpt.insert(0, '<Enter> updates, Add adds new entry')
inpt.grid(row=1, column=0)
# pressing return or left mouse dblclick updates edited line
inpt.bind('<Return>', setList)
inpt.bind('<Double-1>', setList)
# button for sorting
sortBtn = tk.Button(root, text='Sort', command=sortList, width=15)
sortBtn.grid(row=2, column=0, sticky=tk.W)
# button for saving
saveBtn = tk.Button(root, text='Save', command=saveList, width=15)
saveBtn.grid(row=3, column=0, sticky=tk.W)
# button to add
addBtn = tk.Button(root, text='Add entry', command=addItem, width=15)
addBtn.grid(row=2, column=0, sticky=tk.E)
# button to delete
delBtn = tk.Button(root, text='Delete Selected', command=deleteItem, width=15)
delBtn.grid(row=3, column=0, sticky=tk.E)
# load the listbox with data
for item in slist:
    lb.insert(tk.END, item)
 
# left mouse click to display selection
lb.bind('<ButtonRelease-1>', getItem)
 
root.mainloop() #start event-response loop
````
