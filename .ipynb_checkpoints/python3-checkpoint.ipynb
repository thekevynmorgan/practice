## Generators
Generator functions allow you to declare a function that behaves like an iterator, i.e. it can be used in a for loop. They use the yield statement whenever they want to return data. Each time `next()` is called on it, the generator resumes where it left off (it remembers all the data values and which statement was last executed).

### A Simple Example of a Generator
At each `yield` statement, the state of the function is preserved. When generators terminate, they automatically raise `StopIteration`

```python
def seq(data):
    for index in range(0, len(data)):
        yield data[index]

for char in seq('golf'):
    print(char)
```

### Generator Expressions
Some simple generators can be coded as expressions. Generator expressions are more compact but less versatile than full generator definitions. The syntax and concept is similar to list comprehensions but uses parentheses instead of square brackets. The main advantage of generator over a list is that it takes much less memory.

```python
list_comp = [x * 5 for x in range(1000)]
gen_exp = (x * 5 for x in range(1000))
print(gen_exp)
from sys import getsizeof
print(getsizeof(list_comp))
print(getsizeof(gen_exp))
```

### Why Generators?
Generators can be though of as returning multiple items, but instead of returning them all at once they return them one-by-one.

Use them when you are not sure that you will need all of the items, or you have limited space resources (embedded systems).

Another use for generators (that is really the same) is to replace `callbacks` with iterations. In some situations you want a function to do a lot of work and occasionally report back to the caller, eg. a file-system search.

The error indicates that the class "does not know" how to add two vectors together. To remedy this we need to overload the `+` operator for the `Vector` class.

```python
class Vector(object):
    def __init__(self, *args):
        """Create a vector"""
        if len(args) < 2:
            self.x = 0
            self.y = 0
        else:
            self.x = args[0]
            self.y = args[1]
    def __str__(self):
        return "(" + str(self.x) + "," + str(self.y) + ")"
    def __add__(self, other):
        """Returns vector addition"""
        return Vector(self.x + other.x, self.y + other.y)

A = Vector(3,2)
B = Vector(1,5)
print(A+B)
    
```

### Vector Magnitude
The length of a vector (displacement distance) can be found using everyone's favourite Pythagoras theorum:
$$ l = \sqrt{x^2 + y^2} $$
We can provide a new method `mag()`

```python
import math

class Vector(object):
    def __init__(self, *args):
        """Create a vector"""
        if len(args) < 2:
            self.x = 0
            self.y = 0
        else:
            self.x = args[0]
            self.y = args[1]
    def __str__(self):
        return "(" + str(self.x) + "," + str(self.y) + ")"
    def __add__(self, other):
        """Returns vector addition"""
        return Vector(self.x + other.x, self.y + other.y)
    def mag(self):
        return math.sqrt(self.x ** 2 + self.y ** 2)

A = Vector(3,2)
B = Vector(1,5)
C = A+B
print(C)
print(C.mag())
```

## Recursion
Recursion defines a solution to a problem in terms of the same problem, typically of smaller size. Recursion makes it possible to express solutions to certain types of problem very concisely and elegantly.

A function is called *recursive* if it makes a call to itself. Typically, a recursive function will have a terminating condition and one or more recursive calls to itself.

## Factorial
The factorial of a positive integer is that number multiplied by all lesser integers down to 1

Thus `factorial(5)` is 5 x 4 x 3 x 2 x 1

Note that `factorial(6) = 6 * factorial(5)`

Note also that we have a stopping condition. `factorial(1) = 1`

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(6))
```

Each state of the execution of the function is stored at each recursive call. The intermediate results are placed on a **stack** until the expression can be fully evaluated.

## Linked Lists
Although Python has its own `List` class, there are circumstances where a list constructed in a slightly different manner might be beneficial. The most common form of dynamic list (lists that can grow or shrink) is implemented as a Linked List. Linked Lists may be *singly* or *doubly* linked.

In a singly linked list, each item is a NODE having a VALUE and a reference to the NEXT node, and the first and last items are referenced in the member fields of the linked list. In a doubly linked list, each NODE also has a reference to the PREVIOUS node. The following code implements a singly linked list.

```python
class Node:
    def __init__(self, value = None, next_value = None):
        self.value = value
        self.next = next_value
    def __str__(self):
        return ('Node [' + str(self.value) + ']')

class LinkedList:
    def __init__(self):
        self.first = None
        self.last = None
    def insert(self, x):
        if self.first == None:
            self.first = Node(x, None)
            self.last = self.first
        elif self.last == self.first:
            self.last = Node(x, None)
            self.first.next = self.last
        else:
            current = Node(x, None)
            self.last.next = current
            self.last = current
    def __str__(self):
        if self.first != None:
            current = self.first
            out = 'LinkedList [\n' + str(current) + '\n'
            while current.next != None:
                current = current.next
                out += str(current) + '\n'
            return out + ']'
        return 'LinkedList []'
    def clear(self):
        self.__init__()

    def remove(self, value):
        current = self.first
        previous = None
        while current is not None:
            if current.value == value:
                if previous is not None:
                    previous.next = current.next
                    if previous.next == None:
                        self.last = None
                else:
                    self.first = current.next
            previous = current
            current = current.next

lst = LinkedList()
lst.insert("Tom")
lst.insert("Dick")
lst.insert("Harry")
print(lst)
lst.remove("Dick")
print(lst)
lst.remove("Tom")
print(lst)
lst.remove("Harry")
print(lst)
```

## Stacks
A stack is an ordered collection of items where the addition of new items and the removal of existing items always takes place at the same end.

This end is commonly referred to as the "top". The end opposite the top is known as the "base".

Let's make a class `Stack` which creates a new stack that is empty. It needs no parameters and returns an empty stack.

It will have the following methods:
* `push(item)` adds a new item to the top of the stack. It needs the item and returns nothing.
* `pop()` removes the top item from the stack. It needs no parameters and returns the item. The stack is modified.
* `peek()` returns the top item from the stack but does not remove it. It needs no parameters, The stack is not modified.
* `isEmpty()` tests to see whether the stack is empty. It needs no parameters and returns a boolean value.
* `size()` returns the number of items on the stack. It needs no parameters and returns an integer.

```python
class Stack:
    def __init__(self):
        self.items = []
    def __str__(self):
        return str(self.items)
    def isEmpty(self):
        return self.items == []
    def push(self, item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def peek(self):
        return self.items[len(self.items) - 1]
    def size(self):
        return len(self.items)

a = Stack()
a.push(2)
a.push(5)
a.push(3)
print(a)
a.pop()
print(a)
```

## Queues
Queues are a common requirement in many solutions to problems. Items enter at the rear of the queue and leave at the front, in strict order. A Keyboard Buffer where keystrokes are stored is a good example of this. Consider also the live-streaming of video content.

Let us make a class `Queue` which creates a new queues that is empty. It needs no parameters and returns an empty queue.
* `enqueue(item)` adds a new item to the rear of the queue. It need the item and returns nothing.
* `dequeue()` removes the front item from the queue. It needs no parameters and returns the item. The queue is modified.
* `isEmpty()` tests to see whether the queue is empty. It needs no parameters and returns a boolean value.
* `size()` returns the number of items in the queue. It needs no parameters and returns an integer.

```python
class Queue:
    def __init__(self):
        self.items = []
    def __str__(self):
        return str(self.items)
    def enqueue(self, item):
        self.items.insert(0, item)
    def dequeue(self):
        return self.items.pop()
    def size(self):
        return len(self.items)
```

## Stopping Criteria
Given a list of `n` elements, arrange the list in non-decreasing order (allow for duplicates). How do we know when this is achieved? We need a **stopping criteria**. In this case:
* For each adjacent pair of elements, if the larger indexed element is greater than or equal to the smaller indexed element, the list is sorted.
* If this is not the case, then we need to change the order in some way.
* A simple way is to swap adjacent elements when the first condition is not true for a given pair.

## Iterating
Consider the list `[8, 7, 6, 5, 4, 3, 2, 1]`

Pseudocode:

For every index i except the last: if `list[i] > list[i+1]` swap `list[i]` and `list[i+1]`

Of course doing this does not guarantee that the list is sorted , but does it guarantee anything?

Yes, after a single pass through the elements, the nth element is guaranteed to be the largest: `[7, 6, 5, 4, 3, 2, 1, 8]`

So we can do it again, reducing the length by one as there is no need to consider the last item.

## Bubblesort

```python
import time

start = time.time()
def bubblesort(aList):
    for i in range(0, len(aList) - 1):
        bubble(aList, i)

def bubble(aList, i):
    for j in range(0, len(aList) - i - 1):
        if aList[j] > aList[j+1]:
            aList[j], aList[j+1] = aList[j+1], aList[j]
    print("Pass: %s" % str(aList))

al = [2, 3, 1, 5, 6, 4]
bubblesort(al)

end = time.time()
print((end-start))
```

### Performance
Bubblesort, as implemented, is an $O(n^2$) algorithm. The number of operations performed and thus the total time taken is proportional to the square of the number of elements $ (n-1) + (n-2) + (n-3) + \dots $

This can be *slightly improved* upon only by making swaps when they are necessary, i.e. detecting when the list is already sorted (when no swap has occurred in the most recent pass).

### Optimisation
Can we reduce the number of swaps?
* At the end of each of the outer loops that we have defined, we guarantee that the last element is the largest. That is **all** that is guaranteed. (This is known as a loop *invariant*)
* Thus we could achieve the same by finding the maximum value in the part of the list not yet sorted and swap it with the last element (of that part).

This gives is the **selection sort** algorithm.

### Selection Sort
This algorithm keeps track of the maximum value as it scans through the list, and it swaps it into place.

```python
import time
start = time.time()

def selection(aList):
    for i in range(len(aList)):
        select(aList, i)

def select(aList, i):
    top = len(aList) - i - 1
    for j in range(top):
        if (aList[j] > aList[top]):
            top = j
    aList[top], aList[len(aList) - i - 1] = aList[len(aList) - i - 1], aList[top]
    print(aList)

al = [1, 3, 5, 7, 2, 4, 0, 6]
selection(al)
print(al)

end = time.time()
print((end-start)*1000)
```

### Performance
Although we have reduced the number of swaps that need to be performed, we have not reduced the number of operations (we still do pair-wise comparisons).

The algorithm thus remains an $O(n^2$) algorithm which does not scale well in terms of performance.

`n = 10` produced 100 operations

`n = 100` produces 10000 operations

## Is there a better way?
There are a number of sorting methods based upon a *divide and conquer* approach. The *Quicksort* algorithm has an advantage over many such algorithms in that the sorting can be performed 'in place' avoiding the need for additional data structures (although conceptually it is a 'tree' based approach). The *Quicksort* algorithm may be described as follows:
1. Select a pivot value
2. Move (if necessary) values which are less than the pivot value to the left of the pivot.
3. Move values which are greater than or equal to the pivot value to the right of the pivot.

We now have two lists, one with values less than the pivot, and one with values greater than the pivot.

Repeat this process for each of the new lists. The stopping condition will occur when there is only one value in the list.

```python
from time import time
s = time()
def partition(myList, start, end):
    pivot = myList[start]
    left = start + 1
    right = end
    while right >= left:
        while left <= right and myList[left] <= pivot:
            left = left + 1
        while myList[right] >= pivot and right >= left:
            right = right - 1
        if right > left:
            myList[left], myList[right] = myList[right], myList[left]
    myList[start], myList[right] = myList[right], myList[start]
    return right

def quicksort(myList, start, end):
    if start < end: #stopping rule
        # partition the list
        pivot = partition(myList, start, end)
        # sort both halves
        quicksort(myList, start, pivot-1)
        quicksort(myList, pivot+1, end)
    return myList
e = time()
print(quicksort([1,3,5,7,9,0,8,6,4,2], start=0, end=9))
print(e-s)
```

### Performance
If we are unlucky, the pivot might be either the smallest or the largest element in the list. One of the partitions will contain no elements and the other partition will contain `n-1 * n-1` elements (all but the pivot). This will result in $O(n^2)$ running time.

When the partitions are as evenly balanced as possible (their sizes are either equal or are within 1 of each other), the list size effectively halves on each iteration. This gives a performance of $O(n \log n)$. Mathematically it can be shown that the average performance of Quicksort is $O(n \log n)$.

## Better Algorithms?
* Heapsort is slightly slower in average performance than quicksort, but has the advantage of an $O(n \log n)$ worst case.
* Early versions of Python used a Quicksort algorithm for list sorting. More recently, this has been replaced with an adaptive mergesort algorithm.
* There are quicksort algorithms that are much improved by intelligent selection of the pivot.

```python
import numpy as np

import matplotlib.pyplot as plt

# Scale this up and down...
scale = 1000
x = np.linspace(0.01, scale, 10000)
y = np.linspace(0.01, scale, 10000)

# Create the plot
plt.figure(figsize=(10, 6))  # Width and height in inches

# Plot the lines
plt.plot(x, y, label="n", linewidth=2)
plt.plot(x, y**2, label="n^2", color="orange", linewidth=2)
plt.plot(x, y * np.log(y), label="n log(n)", color="green", linewidth=2)

# Set limits
plt.xlim(0, 100)
plt.ylim(0, 1000)

# Add title and legend
plt.title("Big O")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

## Measuring Performance
The measurement of code performance is termed *profiling*.

Python has two profiling modules: `profile` and `cProfile`.

Much can be achieved with simple timing functions. By obtaining the start and end times of a function call we can estimate execution speed. This can be performed on variably sized lists, for example, to measure actual implementations across different scales.

```python
from random import randint
from time import time

import numpy as np

import matplotlib.pyplot as plt

def bubblesort(aList):
    for i in range(0, len(aList) - 1):
        bubble(aList, i)

def bubble(aList, i):
    for j in range(0, len(aList) - i - 1):
        if aList[j] > aList[j+1]:
            aList[j], aList[j+1] = aList[j+1], aList[j]
            
def partition(myList, start, end):
    pivot = myList[start]
    left = start + 1
    right = end
    while right >= left:
        while left <= right and myList[left] <= pivot:
            left = left + 1
        while myList[right] >= pivot and right >= left:
            right = right - 1
        if right > left:
            myList[left], myList[right] = myList[right], myList[left]
    myList[start], myList[right] = myList[right], myList[start]
    return right

def quicksort(myList, start, end):
    if start < end: #stopping rule
        # partition the list
        pivot = partition(myList, start, end)
        # sort both halves
        quicksort(myList, start, pivot-1)
        quicksort(myList, pivot+1, end)
    return myList

times = []
jump = 200
test_max = 5000

for i in range(0, test_max, jump):
    l = []
    for j in range(1, 10+i):
        l.append(randint(0,99))
    
    # Quicksort
    start = time()
    quicksort(l, 0, len(l)-1)
    stop = time()
    qst = stop-start
    
    # Bubblesort
    start = time()
    bubblesort(l)
    stop = time()
    bst = stop-start

    times.append([qst,bst])

plt.plot(times)
plt.ylabel('time(s)')
plt.draw()
plt.show()
```
